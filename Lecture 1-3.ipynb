```python
def lf17(A,B,C):
  if (A>B and B>C) or (A<B and B<C):
    nA = A*2
    nB = B*2
    nC = C*2
  else:
    nA = -A
    nB = -B
    nC = -C
  return nA, nB, nC

lf17(1,2,3)
lf17(7,3,2)
lf17(2,1,6)
```
